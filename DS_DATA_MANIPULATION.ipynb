#### GROUP BY

```python
#Import libraries
from urllib.request import urlretrieve
import pandas as pd

# Assign url of file: url
url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'

# Save file locally
urlretrieve(url, r'C:\Users\Mumes\Desktop\owid-covid-data.csv')

df = pd.read_csv(r'C:\Users\Mumes\Desktop\owid-covid-data.csv')
# Agregamos una columna con el nÃºmero de mes
df['month'] = pd.DatetimeIndex(df['date']).month
#filtramos por paÃ­ses
lista = ['Denmark','Ireland','Argentina', 'Brazil', 'Chile', 'Peru','Mexico', 'United States', 'France', 'Italy', 'Spain', 'United Kingdom', 'India', 'China', 'South Korea', 'Australia', 'New Zealand']
df_new = df[df['location'].isin(lista)]
#eliminamos el mes 12 por no ser relevante
lista_mes = [12]
df_final = df_new[~df_new['month'].isin(lista_mes)]

#agrupamos
groupby_location = df_final.groupby(['continent','location', 'month']).agg({'new_cases':'sum', 'new_deaths':'sum', 'population':'max'})  
groupby_location['%death_cases'] = groupby_location['new_deaths'] / groupby_location['new_cases']
groupby_location['%cases_pop'] = groupby_location['new_cases'] / groupby_location['population']
groupby_location['%death_pop'] = groupby_location['new_deaths'] / groupby_location['population']
#exportamos
groupby_location.to_excel(r'C:\Users\Mumes\Desktop\groupby_location.xlsx')
```
