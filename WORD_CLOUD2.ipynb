```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
#df1 = pd.read_csv(r'C:\Users\Mumes\Desktop\Chat de Whatsapp con Lu Manu.txt', sep=r'[0-9] -', names=['message','time'], engine='python')
df1 = pd.read_csv(r'C:\Users\Mumes\Desktop\Chat de Whatsapp con FMS.txt', sep=r'[0-9] -', names=['message','time'], engine='python')
```

```python
from stop_words import get_stop_words

stop_words = get_stop_words('spanish')

lista =["Yo","Como", "Pero"]

stop_words1 = stop_words + lista
```

```python
#Split the message further into: name and original message:
df2 = df1['message'].str.split(":", expand=True,n=1)
df_all = pd.concat([df1, df2], axis=1)
df_all = df_all.rename(columns={'message': 'total', 0:'name', 1:'message'})
df_all.drop('total', axis=1, inplace=True)

#Saving the messages which are in the time column instead of message column
df_all.loc[df_all.time.str.contains(r'[a-zA-Z]')==True, 'message'] = df_all[df_all.time.str.contains(r'[a-zA-Z]')==True].time
df_all.fillna(' ', inplace=True)

#Delete rows where name includes an activity on group #
df_all = df_all[df_all.name.str.contains("added|changed|created|left")==False]

#Store the text in a variable
text = ' '.join(df_all['message'])


#stop_words = ["Pero","en","Bueno","me"," ","no","lo","ya","vos","esta", "con", "se", "Pero", "el", "Si", "Yo", " Yo", "estoy", "la", "de", "que", "para", "un", "es", "mi", "por", "una", "ma", "te"]



# Create and generate a word cloud image:
wordcloud = WordCloud(stopwords = stop_words1, max_words=50).generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#wordcloud.to_file(r'C:\Users\Mumes\Desktop\img_Lu.png')
wordcloud.to_file(r'C:\Users\Mumes\Desktop\img_fms.png')
```
