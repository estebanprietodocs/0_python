```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\Users\Mumes\Documents\ds_salary_proj\salary_data_cleaned.csv')
df.head(2)
```

```python
# Start with one review:
text = df['Job Description'][0] ### texto de la posicion 0

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file(r'C:\Users\Mumes\0_python\texto_posicion_0.png')
```
